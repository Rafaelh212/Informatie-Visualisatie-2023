## Datasets and Preprocessing

### Dataset 1 

**Title**: Substances risk factor vs direct deaths  
**Source URL**: https://www.kaggle.com/datasets/programmerrdai/drugs-use?resource=download&select=substances-risk-factor-vs-direct-deaths.csv  
**Description**: The dataset is about death by drug or alcohol. It contains records of countries between 1990 and 2019. The dataset looks at which substances cause deaths, disorders and risks in each country in each year from 1990 to 2019. With a quick look one can notice that the number of deaths increase per year in most countries. This is probably due to the advanced progress in technology and transportation which makes it easier to access drugs and alcohol.  

### Dataset 2 
**Title**: Drug use by age  
**Source URL**: https://www.kaggle.com/datasets/tunguz/drug-use-by-age  
**Description**: This dataset looks at different age groups and the usage of different drugs. It goes from the age of 12 to 65. This dataset is from an experiment conducted on 55268 people ranging between these ages. It covers the usage of 13 different drugs. The dataset looks at the percentage of age groups that used the drug in the past 12 months and the median frequency with which they used the drug in the past 12 months

### Dataset 3 
**Title**: Drug Use, Employment, Work Absence, Income, Race, Education (NSDUH Workforce Adults.csv)  
**Source URL**: https://data.world/balexturner/drug-use-employmen t-work-absence-income-race-education  
**Description**: This dataset lists adults with their employment, work absence, income, race and education and shows which kinds of drugs they have interacted with before. This data was collected in the national survey of drug use and health conducted in 2015. This dataset could be used to find correlations between drug use and employment or between demographic factors and drug use.

### Argument 2 
Figure 2 shows the deaths caused by illicit drugs vs the deaths caused by alcohol in each country or region in the past 19 years. Illicit drugs are drugs prohibited by international drug control treaties (Ritchie, H., & Roser, M., 2022). One can argue that with the legalization of these ilicit drugs the death toll will increase to that of alcohol or past that of alcohol which is highly unwanted. Moreover when observing figure 2 countries which are wealthy tend to have more deaths meaning helping people out of poverty doesn't help with preventing death by illicit drugs and might increase it. This is why to be able to bring people out of poverty while maintaining or reducing the amount of deaths caused by drugs the best plan of action is to heavily prohibit and ban the use of illicit drugs. This can be further proven by the fact that arab countries, where the use of drugs will be heavily punished has less deaths compared to countries with less hefty penalties. 

```python
import pandas as pd
import plotly.express as px

aggregated = pd.read_csv('drug-use-by-age.csv')

aggregated.replace('-', 0, inplace=True)

aggregated.to_csv('aggregated_drug_use_by_age.csv')

aggregated['cocaine-frequency'] = pd.to_numeric(aggregated['cocaine-frequency'], errors = 'coerce')
aggregated['crack-frequency'] = pd.to_numeric(aggregated['crack-frequency'], errors = 'coerce')
aggregated['heroin-frequency'] = pd.to_numeric(aggregated['heroin-frequency'], errors = 'coerce')
aggregated['meth-frequency'] = pd.to_numeric(aggregated['meth-frequency'], errors = 'coerce')

marijuana_frequency_sum = aggregated['marijuana-frequency'].sum()
cocaine_frequency_sum = aggregated['cocaine-frequency'].sum()
crack_frequency_sum = aggregated['crack-frequency'].sum()
heroin_frequency_sum = aggregated['heroin-frequency'].sum()
hallucinogen_frequency_sum = aggregated['hallucinogen-frequency'].sum()
meth_frequency_sum = aggregated['meth-frequency'].sum()

frequencies = [marijuana_frequency_sum, cocaine_frequency_sum, crack_frequency_sum,
               heroin_frequency_sum, hallucinogen_frequency_sum, meth_frequency_sum]
text = ['Marijuana', 'Cocaine', 'Crack', 'Heroin', 'Hallucinogen', 'Meth']
df = pd.DataFrame(dict(frequencies = frequencies))

fig = px.pie(df, values = 'frequencies', names = text,
             title = 'Figure 3: Proportions of the median numbers of times recreational drugs have been used in 12 months')

fig.update_traces(textposition = 'outside', textinfo='label, percent', marker = dict(colors = px.colors.qualitative.T10))
fig.update_layout(showlegend = False)
fig.show()
```

### Argument 3
Figure 3 shows the ratio between the amount of times adults and minors used six kinds of recreational drugs during the span of 12 months. It shows that 70,21% of the drugs used are hard drugs, namely cocaine, crack, heroin and meth. Some countries decriminalized the use of soft drugs like marijuana and hallucinogen. For example, The Netherlands allow the usage of soft drugs and state that the risks soft drugs are less serious because soft drugs are less dangerous to the health than hard drugs (Government of the Netherlands, 2023). Since hard drugs are the most used kind of drug and very hazardous for the health, all drugs should be illegal and never be decriminalized.
